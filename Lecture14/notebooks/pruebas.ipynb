# NVIDIA NIM API — Guía de configuración

## 1. Crear cuenta y obtener la API Key

1. Ir a [build.nvidia.com/models](https://build.nvidia.com/models)
2. Hacer clic en **Login** (esquina superior derecha)
3. Crear una cuenta en el **NVIDIA Developer Program** (gratuito) o iniciar sesión si ya tienes una
4. Una vez autenticado, hacer clic en tu perfil → **Get API Key** (o el botón aparece al seleccionar cualquier modelo con *Free Endpoint*)
5. Copiar la key — comienza con `nvapi-`

> ⚠️ Guarda la key de inmediato, NVIDIA no la vuelve a mostrar completa.

---

## 2. Configurar variables de entorno

Crear un archivo `.env` en la raíz del proyecto:

```
NVIDIA_API_KEY=nvapi-XXXXXXXXXXXXXXXXXXXX
```

Agregar `.env` al `.gitignore`:

```
echo ".env" >> .gitignore
```

pip install openai python-dotenv

````python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()

NVIDIA_API_KEY = os.getenv("NVIDIA_API_KEY")
````

````python
client = OpenAI(
    base_url="https://integrate.api.nvidia.com/v1",
    api_key=NVIDIA_API_KEY  # desde build.nvidia.com/models
)
````

````python
models = client.models.list()
for m in models.data:
    print(m.id)
````

````python
response = client.chat.completions.create(
    model="minimaxai/minimax-m2.7",
    messages=[{"role": "user", "content": "Explica backpropagation en 3 líneas"}],
    temperature=0.6,
    max_tokens=512,
)

print(response.choices[0].message.content)
````
